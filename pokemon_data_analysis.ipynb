Making Changes to the Data
# Adding a Total column. 

```python
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Speed']

#Similar Method => df['Total'] = df.iloc[:, 4:10].sum(axis=1)

df.head(5)
```

#Changing the order of the DataFrame columns. 
#In this case, changing the position of the Tota column

```python
cols = list(df.columns.values)
df = df[['#', 'Name', 'Type 1','Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk','Sp. Def','Speed', 'Total', 'Generation', 'Legendary' ]]
# similar to indexing with range : df = df[cols[0:4]+ [cols[-1]]+cols[4:12]]
df.head(5)
```

Aggregate Statistics(Groupby)

```python
df = pd.read_csv("Updated.csv")
df
```

```python
#Checking the average of all the Type 1 pokemons
df.groupby(['Type 1']).mean()
```

```python
#Sorting values 
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
#Counting the pokemons in Type 1 column and getting only the count column 
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```
